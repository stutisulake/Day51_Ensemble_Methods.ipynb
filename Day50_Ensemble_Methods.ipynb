# Ensemble Methods

## Work Summary

• Today, I advanced from optimizing individual models to mastering Ensemble Methods.

• I learned how to combine multiple learners to create a more robust and accurate strong learner.

• I explored the theoretical foundation of how aggregating diverse models with different biases and strengths can cancel out individual errors and improve generalization.

• I implemented both Voting Classifiers (Hard and Soft).

• I also implemented Stacking Classifiers where a meta-learner is trained to intelligently weight predictions of base models.

• Finally, I performed a comparison to determine whether the increased computational complexity of ensemble methods was justified by the performance gains.


## Learnings / Outcomes

• Weak vs Strong Learners  
I understood the Wisdom of the Crowd concept where multiple weak learners combine to form a strong learner.

• Model Diversity  
I learned that ensemble models perform better when base models are different such as Logistic Regression and Decision Tree.

• Hard Voting  
I learned that hard voting uses majority voting to predict the final class label.

• Soft Voting  
I learned that soft voting averages prediction probabilities and usually gives better performance because confident predictions get higher importance.

• Stacking Logic  
I implemented stacking where predictions from base models are used as input features for a meta-learner to improve final predictions.


## Blockers / Risks

• The Probability Trap  
I encountered an error during soft voting because one model did not support probability prediction by default. I solved this by enabling probability=True for SVM models.

• Diminishing Returns  
I observed that adding too many models did not significantly improve accuracy but increased training time.

• Overfitting in Stacking  
I noticed that complex meta-learners can overfit base model predictions, so I used Logistic Regression as a simpler final estimator.
